```
house_pts = np.array([[0, 0, 1, 1, 0, 0, 1, 1, 0.5],
                     [0, 1, 1, 0, 0, 1, 1, 0, 0.5],
                     [0, 0, 0, 0, 1, 1, 1, 1, 1.5]])
one_rows = np.ones((1, 9))
house_mat = np.concatenate((house_pts, one_rows))
print(house_mat)
```

```
M_int = np.array([[-1,    0, 2],
                  [0.  , -1, 2],
                  [0   ,    0,   1]])

# VIEW 1
# R_view_1 = np.array([[0.5, -0.5, 0.7071],
#                     [0.8535, 0.1464, -0.5],
#                     [0.1464, 0.8535, 0.5]])
# t_view_1 = np.array([[10],
#                      [10],
#                      [10]])
R_view_1 = np.array([[0.707, 0.707, 0],
                  [-0.707, 0.707, 0],
                  [ 0    , 0    , 1]])
t_view_1 = np.array([[-3],
                    [-0.5],
                    [3]])
M_ext_view_1 = np.concatenate((R_view_1, t_view_1), axis = 1)

# VIEW 2
# R_view_2 = np.array([[0.75, -0.433, 0.5],
#                     [0.625, 0.2165, -0.75],
#                     [0.2165, 0.875, 0.433]])
# t_view_2 = np.array([[10],
#                      [10],
#                      [10]])
R_view_2 = np.array([[0.866, -0.5, 0],
                    [0.5, 0.866, 0],
                    [ 0    , 0    , 1]])
t_view_2 = np.array([[3],
                    [5],
                    [3]])
M_ext_view_2 = np.concatenate((R_view_2, t_view_2), axis = 1)

# VIEW 1
mat_view_1 = np.matmul(np.matmul(M_int, M_ext_view_1), house_mat)
result_mat_view_1 = np.array([mat_view_1[0] / mat_view_1[2], mat_view_1[1] / mat_view_1[2]]).T

# VIEW 2
mat_view_2 = np.matmul(np.matmul(M_int, M_ext_view_2), house_mat)
result_mat_view_2 = np.array([mat_view_2[0] / mat_view_2[2], mat_view_2[1] / mat_view_2[2]]).T

print(result_mat_view_1)
print()
print(mat_view_1)
print()
print(result_mat_view_2)
```

```
draw_my_object(result_mat_view_1)
draw_my_object(result_mat_view_2)
draw_my_3d_object(house_pts)
```

```
# STEP 2: From image pixels, multiple by K inverse to get rays in camera coordinates (think of these rays as directional light detectors)

view1_ray = np.matmul(np.linalg.inv(M_int), mat_view_1)
view2_ray = np.matmul(np.linalg.inv(M_int), mat_view_2)

print(view1_ray)
print()
print(view2_ray)
```

```
# STEREO RECONSTRUCTION 
temp_rows = np.array([[0,0,0,1]])
T1w = np.append(M_ext_view_1,temp_rows,axis=0)
T2w = np.append(M_ext_view_2,temp_rows,axis=0)

Tw1 = np.linalg.inv(T1w) 
Tw2 = np.linalg.inv(T2w)

T12 = np.matmul(T1w,Tw2)
print('Rotation from view 2 to view 1 coordinate frame')
print(T12)

R12 = T12[0:3,0:3] 
t12 = T12[0:3,3]

print('Rotation from view 2 to view 1')
print(R12)
print('Translation from view 2 to view 1')
print(t12)
```

```
def triangulate_dlt(view1_ray, view2_ray, R12,t12):
  outpoints = np.array([[],
                        [],
                        []])
  for i in range(9):
    view1_point = view1_ray[:,i].reshape(3,1)
    view2_point = view2_ray[:,i].reshape(3,1)
    temp = np.matmul(R12, view2_point)
    
    q = view1_point*temp
    normalize = q / np.linalg.norm(q)
    A = np.append(np.append(view1_point, temp ,axis=1), q,axis=1)
    
    solveit = np.matmul(np.linalg.inv(A),t12)
    
    
    point_1 = solveit[0]*view1_point + 0.5*solveit[2]* normalize
    
    # point_1 = np.asarray(point_1)

    point_1 = np.append(point_1,[[1]],axis=0)
    
    outpoints = np.append(outpoints,np.matmul(M_ext_view_1,point_1),axis=1)
  return outpoints
```

```
world_points = triangulate_dlt(view1_ray, view2_ray, R12,t12)
print(world_points)
```

```
row_ones = np.ones((1,9))
world_points = np.concatenate((world_points, row_ones))

mat_view_1 = np.matmul(np.matmul(M_int, M_ext_view_1), world_points)
result_mat_view_1 = np.array([mat_view_1[0]/mat_view_1[2],mat_view_1[1]/mat_view_1[2]]).T

draw_my_object(result_mat_view_1)
draw_my_3d_object(world_points)

print('Reprojection error')
print((world_points - house_mat))
```
