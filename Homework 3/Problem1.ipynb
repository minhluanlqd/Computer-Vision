```
import matplotlib.pyplot as plt
import numpy as np
```

```
house_pts = np.array([[0, 0, 1, 1, 0, 0, 1, 1, 0.5],
                     [0, 1, 1, 0, 0, 1, 1, 0, 0.5],
                     [0, 0, 0, 0, 1, 1, 1, 1, 1.5]])
print(house_pts)
```

```
one_rows = np.ones((1, 9))
house_mat = np.concatenate((house_pts, one_rows))
print(house_mat)
```

```
def draw_my_object(final):
  p1 = [final[0][0], final[0][1]]
  p2 = [final[1][0], final[1][1]]
  p3 = [final[2][0], final[2][1]]
  p4 = [final[3][0], final[3][1]]
  p5 = [final[4][0], final[4][1]]
  p6 = [final[5][0], final[5][1]]
  p7 = [final[6][0], final[6][1]]
  p8 = [final[7][0], final[7][1]]
  p9 = [final[8][0], final[8][1]]

  plt.figure()
  x = [p1[0], p2[0], p3[0], p4[0], p5[0], p6[0], p7[0], p8[0], p9[0]]
  y = [p1[1], p2[1], p3[1], p4[1], p5[1], p6[1], p7[1], p8[1], p9[1]]

  plt.scatter(x[0], y[0], color="red") #origin
  plt.scatter(x[8], y[8], color="blue") # rooftop point

  line1 = plt.Line2D([p1[0], p2[0]], [p1[1], p2[1]], color="black")
  line2 = plt.Line2D([p2[0], p3[0]], [p2[1], p3[1]], color="black")
  line3 = plt.Line2D([p3[0], p4[0]], [p3[1], p4[1]], color="black")
  line4 = plt.Line2D([p4[0], p1[0]], [p4[1], p1[1]], color="black")
  line5 = plt.Line2D([p5[0], p6[0]], [p5[1], p6[1]], color="black")
  line6 = plt.Line2D([p6[0], p7[0]], [p6[1], p7[1]], color="black")
  line7 = plt.Line2D([p7[0], p8[0]], [p7[1], p8[1]], color="black")
  line8 = plt.Line2D([p8[0], p5[0]], [p8[1], p5[1]], color="black")
  line9 = plt.Line2D([p1[0], p5[0]], [p1[1], p5[1]], color="black")
  line10 = plt.Line2D([p2[0], p6[0]], [p2[1], p6[1]], color="black")
  line11 = plt.Line2D([p3[0], p7[0]], [p3[1], p7[1]], color="black")
  line12 = plt.Line2D([p4[0], p8[0]], [p4[1], p8[1]], color="black")
  line13 = plt.Line2D([p6[0], p9[0]], [p6[1], p9[1]], color="black")
  line14 = plt.Line2D([p7[0], p9[0]], [p7[1], p9[1]], color="black")
  line15 = plt.Line2D([p8[0], p9[0]], [p8[1], p9[1]], color="black")
  line16 = plt.Line2D([p5[0], p9[0]], [p5[1], p9[1]], color="black")

  plt.gca().add_line(line1)
  plt.gca().add_line(line2)
  plt.gca().add_line(line3)
  plt.gca().add_line(line4)
  plt.gca().add_line(line5)
  plt.gca().add_line(line6)
  plt.gca().add_line(line7)
  plt.gca().add_line(line8)
  plt.gca().add_line(line9)
  plt.gca().add_line(line10)
  plt.gca().add_line(line11)
  plt.gca().add_line(line12)
  plt.gca().add_line(line13)
  plt.gca().add_line(line14)
  plt.gca().add_line(line15)
  plt.gca().add_line(line16)

  plt.show()
```

```
# LEFT CAMERA
R_view_1 = np.array([[0.707, 0.707, 0],
                  [-0.707, 0.707, 0],
                  [ 0    , 0    , 1]])
t_view_1 = np.array([[-3],
                    [-0.5],
                    [3]])
M_ext_view_1 = np.concatenate((R_view_1, t_view_1), axis = 1)

M_int_view_1 = np.array([[-100,    0, 200],
                      [0.  , -100, 200],
                      [0   ,    0,   1]])
print(M_ext_view_1)
print(M_int_view_1)

result_mat_view_1 = np.matmul(np.matmul(M_int_view_1, M_ext_view_1), house_mat)
x_y_mat_view_1 = np.array([result_mat_view_1[0]/result_mat_view_1[2], result_mat_view_1[1]/result_mat_view_1[2]])
final_view_1 = x_y_mat_view_1.T

print(final_view_1)
```
