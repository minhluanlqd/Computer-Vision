```
# Rank of the matrix = maximum number of linearly independent column vectors
# Rank of the matrix = maximum number of linearly independent row vectors
from numpy.linalg import matrix_rank
from numpy.linalg import svd
```

```
mat = [[-1.32, -0.18, 2.13],
       [2.64, -4.68, 4.65],
       [1.47, -4.75, 6.80]]
# mat2 = [[1, 2, 3],
#         [2, 4, 6],
#         [3, 5, 8]]
print(mat)
```

```
u, sigma, vh = svd(mat, full_matrices=True)
# u2, sigma2, vh2 = svd(mat2, full_matrices=True)
print(u)
print("\n")
print(sigma)
print("\n")
print(vh)
# print(u2)
# print("\n")
# print(sigma2)
# print("\n")
# print(vh2)
```

```
print("Rank of matrix is: ",  matrix_rank(mat))
```

- Sigma "matrix" has the last value is 0.099.. => Not small enough to consider it as 0. So the Rank of given matrix is 3
- Rank = 3 so there are ***3 linearly independent columns***
