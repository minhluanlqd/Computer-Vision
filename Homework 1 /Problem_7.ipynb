```
import numpy as np
from numpy.linalg import svd
import matplotlib.pyplot as plt
```

```
# Create X and Y coordinates for ellipse
# TODO: Play around with these values to understand what each variable means

a, b = 5, 100 

t = np.arange(0, 100, 0.01)

x = a*np.sin(t)
y = b*np.cos(t)
```

```
# Display the ellipse
plt.plot(x, y, '*-')
plt.show()
```

```
# Rotate the points of the ellipse
cc = np.cos(np.pi / 3)
ss = np.sin(np.pi / 3)

R = np.array([[cc, ss],
              [-ss, cc]])

pts = np.stack((x, y))

rpts = np.dot(R, pts)
```

```
rpts.shape
```

```
u, sigma, vh = svd(rpts, full_matrices=True)
```

```
u
```

```
# TODO: Compute the rotated angle in radians and degrees from rpts
vec1 = np.array([-0.86609104, -0.4998863])
vec2 = np.array([-0.4998863, 0.86609104])
RotAngle = np.arctan(vec1/vec2)

print("Rotated angle in radians from rpts (in view of MAJOR axis): ", RotAngle[0])
print("Rotated angle in degrees from rpts (in view of MAJOR axis): ", np.degrees(RotAngle[0]))
print("Rotated angle in radians from rpts (in view of MINOR axis): ", RotAngle[1])
print("Rotated angle in degrees from rpts (in view of MINOR axis): ", np.degrees(RotAngle[1]))
```

```
# TODO: Change the rotation angle and repeat computation
est_rot_angle = np.pi / 4
```

```
# Rotate original points based on estimate
cc = np.cos(est_rot_angle)
ss = np.sin(est_rot_angle)

R = np.array([[cc, ss],
              [-ss, cc]])

pts = np.stack((x, y))

est_rpts = np.dot(R, pts)

# Display results
plt.plot(rpts[0,:], rpts[1,:], 'b*', label="pi/3")
plt.plot(est_rpts[0,:], est_rpts[1,:], 'r*', label="pi/4")
# TODO: add legend
plt.legend()
plt.show()
```
